# **Bài tập 4**: Vẽ biểu đồ top 20 bài hát có lượng play time cao nhất ở Mỹ (Region == "us")

```python
#get sum stream by track name for us region 
stream_in_us = data_df[data_df['Region']=='us'].groupby(['Track Name'])['Streams'].agg(['sum'])
stream_in_us.rename(columns={'sum': 'total stream in us'}, inplace=True)
stream_in_us
```

```python
#join stream_in_us with feature_df, get only track name available in feature_df
play_time_df = feature_df.set_index(["name"]).join(stream_in_us, how="left")
play_time_df["play time"] = play_time_df["duration"].values * play_time_df["total stream in us"].values
play_time_df
```

```python
top_20_song = play_time_df.nlargest(20, "play time")
ax = top_20_song.plot.barh(use_index=True, y='play time')
```
